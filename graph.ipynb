```python
# Import the pandas library with the usual "pd" shortcut
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

plotdata = pd.DataFrame(
    {"w sentiments": [95, 94, 89, 88],
     "w/o sentiments": [92, 93, 88, 87]
    }, 
    index=["CNN (BERT embedding)", "CNN (Wor2vec embedding)", "LSTM (Wor2vec embedding)", "LSTM (BERT embedding)"])
    
# Plot a bar chart
ax =plotdata.plot(kind="barh")
for container in ax.containers:
    ax.bar_label(container)
ax.legend(loc='upper center')    
plt.title("Performance of the models")
plt.xticks(rotation=30, horizontalalignment="center")
plt.xlabel("F1-score accuracy (%)")
plt.savefig('graph.png')
```
